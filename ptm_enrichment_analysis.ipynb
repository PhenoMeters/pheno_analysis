```python
import pandas as pd
import numpy as np 
import seaborn as sns
```

```python
interfaces_data = pd.read_csv("/people/imal967/git_repos/pheno_analysis/results/merged_interfaces_FINAL2.csv")
interfaces_data
```

```python
pockets_data = pd.read_csv("/people/imal967/git_repos/pheno_analysis/results/merged_pockets_FINAL.csv")
pockets_data
```

```python
# isolating to only inside pocket or interface
inside_pocket_df = pockets_data[pockets_data['inside_pocket'] == 1]
inside_interface_df = interfaces_data[interfaces_data['inside_interface'] == 1]
print(pockets_data.shape)
print(inside_pocket_df.shape)
print(interfaces_data.shape)
print(inside_interface_df.shape)
```

```python
# condensing df's to 1 row per protein and add up the inside pocket
agg_method = {
        "ACC_ID": lambda x: x.iloc[0],
        "GENE": lambda x: x.iloc[0],
        "PROTEIN": lambda x: x.iloc[0],
        "res_number": lambda x: ', '.join(x.astype("string").values),
        "AA": lambda x: ','.join(x.values),
        "closest_pocket": lambda x: ','.join(x.str.slice(1)),
        "inside_pocket": lambda x: x.sum()
    }
inside_pocket_df = inside_pocket_df.groupby(by="ACC_ID", as_index=False).agg(agg_method)
inside_pocket_df
```
