```python
uniprot_list = []
binding_res = []
binding_data = []
active_res = []
active_type = []
DNA_binding_res = []

unitprot_table = pd.read_csv()
for index, row in unitprot_table.iterrows():
    protein_object = am.Protein.from_uniprot_row(row)
    df = pd.DataFrame(protein_object.unravel_sites())
    binding_res.append(df[df["binding"]]["residue_number"].to_list())
    binding_data.append(df[df["binding"]]["binding_data"].to_list())
    active_res.append(df[df["active"]]["residue_number"].to_list())
    active_type.append(df[df["active"]]["active_data"].to_list())
    DNA_binding_res.append(df[df["dna_binding"]]['residue_number'].to_list())

binding_active_dna_binding = pd.DataFrame({'uniprot': uniprot_list, 'active_site_residues': active_res, 'active_site_description': active_type, 'binding_site_residues':binding_res, 'binding_site_description':binding_data, 'dna_binding_residues':DNA_binding_res})
binding_active_dna_binding.to_csv("/people/imal967/git_repos/pheno_analysis/binding_active_dna.csv")
```
